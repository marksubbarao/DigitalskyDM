### Let's get started - Step #1 Acquiring the GAMA Data from the database

We want to get our galaxy catalog from the GAMA survey, to do that we need to query its database. A survey database typically contains several different tables each measuring a number of different quantities. Looking at the structure of the GAMA database (its schema) I see that the quantities I are in two tables, the EMLines table has the coordinates we need to locate objects in 3D, these are sky position (RA and dec) and the redshift(z) which we will use to calculate the distance to the galaxy. That table also contains a measurement of the star formation rate (SFR) and a classification of emmission line galaxies (Emline_Class). There is another table called StellarMasses which has several interesting parameters, we will get  the mass of the galaxy (logmstar), its instrinsic brightness (absMag_r), its age (logage), mass to light ratio (logMoverM_i) and its metalicity (metal).

The database can be accessed using the Structured Query Language(SQL). SQL has a SELECT/FROM/WHERE format, after SELECT you list what parameters you want to pull from the dataase, after WHERE you specifiy which database tables they are located in, and after WHERE you can place conditions on which objects you wish to get back.

So our query will be:

   SELECT  EmLinesPhys.ra, EmLinesPhys.dec, EmLinesPhys.Z_TONRY,
           EmLinesPhys.SFR,EmLinesPhys.EMline_Class,
           StellarMasses.logmstar, StellarMasses.metal, StellarMasses.absMag_r,
           StellarMasses.logage, StellarMasses.logMoverl_i<br>
   FROM    EmLinesPhys JOIN StellarMasses ON EmLinesPhys.CATAID = StellarMasses.CATAID<br>
   WHERE   EmLinesPhys.NQ>2 <br>

The purpose of the JOIN syntax in the FROM clause is to make sure we are looking at the same galaxies in each table. The conditions in the WHERE clause make sure that the galaxies spectrum was high quality and that the redshift was certain.

Using the <a href="http://astroquery.readthedocs.org/en/latest/#">astroquery module</a> we can query the GAMA database inside python and return the results in a data table.

```python
#Imports
from astroquery.gama import GAMA
from astropy.table import Table,Column
from astropy import units as u
from astropy.coordinates import ICRS, CartesianRepresentation, SkyCoord, Distance
import numpy as np
```

```python
#Acquire Data from GAMA database
galaxyTable = GAMA.query_sql('\
   SELECT  EmLinesPhys.ra, EmLinesPhys.dec, EmLinesPhys.Z_TONRY,\
           EmLinesPhys.SFR,EmLinesPhys.EMline_Class,\
           StellarMasses.logmstar, StellarMasses.metal, StellarMasses.absMag_r,\
           StellarMasses.logage, StellarMasses.logMoverl_i\
   FROM    EmLinesPhys JOIN StellarMasses ON EmLinesPhys.CATAID = StellarMasses.CATAID\
   WHERE   EmLinesPhys.NQ>2 '\
)
```

OK, let's now check what the full list of values in the Emline_Class are 

```python
names = set(galaxyTable['EMline_Class'])
names
```

Good, we'll preplace those codes with integers to make coloemaps easier: 1- star forming galaxies, 2 = AGN, non emmission line galaxies are 0

```python
#galaxyTable.remove_column('class')
galaxyTable.add_column(Column(name='class',dtype=int,data=np.zeros(len(galaxyTable))))
for i,EMtype in enumerate(galaxyTable['EMline_Class']):
    if EMtype.strip()=="SF":
        galaxyTable['class'][i]=1
    if EMtype.strip()=='AGN':
        galaxyTable['class'][i]=2 
galaxyTable.remove_column('EMline_Class')
```

### Step #2 Calculate distance to the galaxies from their redshift

We will plot the galaxies in co-moving distance. Now to calculate that we'll need to choose a set of Cosmological parameters. One of the really nice features of astropy is  the built in set of standard cosmologies. Here I choose the Planck13 cosmology defined <a href="http://arxiv.org/abs/1303.5076">here</a>. The redshift Z_TONRY that we've chosen already attempts to correct for the local flow which is nice.

```python
#galaxyTable.remove_column('Distance')
from astropy.cosmology import Planck13
#Calculate a new column of comoving distance to the galaxies
distCol=Column(Planck13.comoving_distance(galaxyTable['Z_TONRY']),\
               name='Dist',meta={'ucd': 'pos.distance'})
#Add that column to the data table (as the third column after ra and dec)
galaxyTable.add_column(distCol,2)
```

```python
#Create a column of coordinate objects because it will be useful later
coordsCol=SkyCoord(galaxyTable['ra'],galaxyTable['DEC'],unit=(u.degree, u.degree),\
                   distance=Distance(galaxyTable['Dist'],u.mpc),frame='icrs')
```

## Write out .speck format for DigitalSky
Now we need to write out the data in the ".speck" format defined by Partiview. This format is a ASCII table where the first three columns are the x,y,z Cartesian coordinates and the remaining columns are whatever additional data we want to carry around. The table begins with a header that names these additional columns.

```python
#Now we'll add the projection into the Galactic coordinate system
galaxyTable.add_column(Column(coordsCol.galactic.represent_as(CartesianRepresentation).x.to(u.Mpc),name='x_gal_coord',meta={'ucd': 'pos.cartesian.x'}),0)
galaxyTable.add_column(Column(coordsCol.galactic.represent_as(CartesianRepresentation).y.to(u.Mpc),name='y_gal_coord',meta={'ucd': 'pos.cartesian.y'}),1)
galaxyTable.add_column(Column(coordsCol.galactic.represent_as(CartesianRepresentation).z.to(u.Mpc),name='z_gal_coord',meta={'ucd': 'pos.cartesian.z'}),2)
```

```python
#Write .speck file for DigitalSky
writeFile = open('GAMA.speck','a')
#Define and write the header
headerText="\
#Galaxy and Mass Assembly galaxy catalog\n\
datavar 0 ra\n\
datavar 1 Dec\n\
datavar 2 Distance\n\
datavar 3 redshift\n\
datavar 4 SFR\n\
datavar 5 logmstar\n\
datavar 6 metalicity\n\
datavar 7 absMag_r\n\
datavar 8 logage\n\
datavar 9 logmasslight\n\
datavar 10 class\n\
texturevar 11\n\
texture -M 1 halo.sgi\n"
writeFile.write(headerText)
#Add Texture column to galaxyTable
galaxyTable['texture']=1
#Write the data
galaxyTable.write(writeFile,format='ascii.no_header')
writeFile.close()
```

### Make .cmap files from colormaps

```python
# coolwarm used for Galaxy Masses
writeFile = open('coolwarm.cmap','w')
ncmap=21
writeFile.write("%d\n" % ncmap)
for i in range(ncmap):
  color=cm.coolwarm(i/(1.0*(ncmap-1.)))
  writeFile.write("%f %f %f\n" % (color[0],color[1],color[2]))
writeFile.close()
```

```python
# gist_rainbow used for Star Formation Rate
writeFile = open('gist_rainbow.cmap','w')
ncmap=41
writeFile.write("%d\n" % ncmap)
for i in range(ncmap):
  color=cm.gist_rainbow(i/(1.0*(ncmap-1.)))
  writeFile.write("%f %f %f\n" % (color[0],color[1],color[2]))
writeFile.close()
```

```python
# spring used for metalicity
writeFile = open('spring.cmap','w')
ncmap=21
writeFile.write("%d\n" % ncmap)
for i in range(ncmap):
  color=cm.spring(i/(1.0*(ncmap-1.)))
  writeFile.write("%f %f %f\n" % (color[0],color[1],color[2]))
writeFile.close()
```

```python
# RdYlBu used for AbsoluteMag
writeFile = open('RdYlBu.cmap','w')
ncmap=21
writeFile.write("%d\n" % ncmap)
for i in range(ncmap):
  color=cm.RdYlBu(i/(1.0*(ncmap-1.)))
  writeFile.write("%f %f %f\n" % (color[0],color[1],color[2]))
writeFile.close()
```

```python
# jet used for Age
writeFile = open('jet.cmap','w')
ncmap=21
writeFile.write("%d\n" % ncmap)
for i in range(ncmap):
  color=cm.jet(i/(1.0*(ncmap-1.)))
  writeFile.write("%f %f %f\n" % (color[0],color[1],color[2]))
writeFile.close()
```

```python
# hot used for M/L
writeFile = open('hot.cmap','w')
ncmap=21
writeFile.write("%d\n" % ncmap)
for i in range(ncmap):
  color=cm.hot(i/(1.0*(ncmap-1.)))
  writeFile.write("%f %f %f\n" % (color[0],color[1],color[2]))
writeFile.close()
```
